```python
df.boxplot(figsize=[20,8])
plt.show()
```

```python
plt.hist(df['Total Volume'],bins=10)
plt.title('Total Volume Histogram')
plt.show()
```

```python
plt.hist(df['Total Bags'],bins=10)
plt.title('Total Bags Histogram')
plt.show()
```

```python
df.plot(kind='density',subplots=True,layout=(6,11),sharex=False,legend=False,fontsize=1,figsize=(18,12))
plt.show()
```

```python
df.iloc[:,0:15]
sns.stripplot(x="region",y="Total Volume",data=df)
plt.title('Region graph')
plt.show()
```

```python
plt.figure(figsize=(15,7))
sns.heatmap(df.corr(),annot=True,linewidths=0.5,linecolor='black',fmt='.2f')
```

```python
df.iloc[:,0:50]
plt.xticks(rotation=90,fontsize=10)
sns.stripplot(x="region",y="Total Volume",data=df)
plt.title('Region Volume',fontsize=10,fontweight="bold")
plt.show()
```
